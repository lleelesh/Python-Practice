```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
sns.stripplot(y='tip',data=tips,jitter=True)
plt.ylabel('tips($)')
plt.show()
```

```python
sns.stripplot(y='tip',data=tips,jitter=True)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.stripplot(y='tip',data=tips,jitter=False)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=tips,size=4,jitter=False)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.swarmplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.swarmplot(x='day',y='tip',data=tips,hue='sex')
plt.ylabel('tip($)')
plt.show()
```

```python
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
plt.subplot(1,3,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.violinplot(x='day',y='tip',data=tips,inner=None,color='lightgray')
sns.stripplot(x='day',y='tip',data=tips,size=4,jitter=True)
plt.ylabel('tip($)')
plt.show()
```
