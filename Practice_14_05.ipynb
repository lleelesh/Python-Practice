```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
fig=plt.figure(figsize=(15,5))
sns.stripplot(y='tip',data=tips,jitter=True)
plt.ylabel('tip($)')
plt.show()
```

```python
fig=plt.figure(figsize=(15,5))
sns.stripplot(y='tip',data=tips,jitter=False)
plt.ylabel('tips($)')
plt.show()
```

```python
fig=plt.figure(figsize=(15,5))
sns.stripplot(x='day',y='tip',data=tips)
plt.ylabel('tips($)')
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=tips,size=4,jitter=False)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=tips,size=4,jitter=True)
plt.ylabel('tip($)')
plt.show()
```

```python
sns.stripplot(x='day',y='tip',data=tips,size=4,jitter=True)
plt.ylabel('tips($)')
plt.xlabel('days')
plt.title('Tips Data')
```

```python
sns.swarmplot(x='day',y='tip',data=tips,size=4)
plt.ylabel('tips($)'),
plt.show(),
```

```python
fig=plt.figure(figsize=(15,4))
sns.swarmplot(x='day',y='tip',data=tips,hue='sex',size=4)
plt.ylabel('tips($)')
plt.show()
```

```python

sns.swarmplot(x='day',y='tip',data=tips,hue='sex',orient='h',size=4)
plt.xlabel('tip($)')
plt.show()
```

```python
fig=plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tips($)')
plt.subplot(1,2,2)
sns.violinplot(x='day',y='tip',data=tips)
plt.ylabel('tips($)')
plt.show()
```

```python
sns.violinplot(x='day',y='tip',data=tips,inner=None,colour='red')
sns.stripplot(x='day',y='tip',data=tips,size=4,jitter=True)
plt.ylabel('tips($)')
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
plt.show()
```
